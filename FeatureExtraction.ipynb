```python
#load stop words
import codecs

from hazm import Normalizer
l = []
with  open('persian-stopwords-master/persian', encoding='utf-8') as f:
    for w in f.readlines():
        l.append(w.replace('\n',''))


len(l)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer( stop_words=l)
X = vectorizer.fit_transform(df[df.columns[-1]] )
```

```python
X.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
from sklearn.decomposition import TruncatedSVD
ncomp=32
svd = TruncatedSVD(n_components=ncomp,algorithm='arpack', random_state=42)
dimX = svd.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
nc=20
losses = []

kmeans = KMeans(n_clusters=nc, random_state=0).fit(dimX)

kmeans.score(dimX)
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([('vectorizer', vectorizer), ('dimenstionality_reduction', svd) , 
                ('clustering' , kmeans)])
```

```python
import random
l=random.sample(list(df.index),50)
pred=pipe.predict([df.iloc[x][df.columns[-1]] for x in l])

for i in range(nc):
    print("-------------------cluster : {}-------------------".format(i))
    for c in range(len(pred)):
        if pred[c]==i:
            print(df.iloc[l[c]][10])
            print("------")
```

```python
# kmeans.cluster_centers_
transformpipe = Pipeline([('vectorizer', vectorizer), ('dimenstionality_reduction', svd)])
# df.apply(lambda )
transformall=transformpipe.transform(df.iloc[:][15])
dffind = df[[0,'cluster']]
dffind[0] = df.index
dffind.head()
```

```python
test=dffind[dffind['cluster']==0][0].values
print(test[:10])
test = sorted(test , reverse=True , key = lambda t : np.linalg.norm(kmeans.cluster_centers_[0]-transformall[t]))
print(test[:10])
```

```python
centroids = {}
for i in range(nc):
    candidate=dffind[dffind['cluster']==i][0].values
    test = sorted(candidate , reverse=True , key = lambda t : np.linalg.norm(kmeans.cluster_centers_[i]-transformall[t]))
    centroids[i] = test[:5]
```

```python
df.iloc[centroids[0][0]][14]
```

```python
%%capture cap --no-stderr
for i in range(nc):
    print("--------------nearest example to centroid {}----------------".format(i))
    for j in centroids[i]:
        print(df.iloc[j][14])
with open('outputcentroid.txt', 'w' , encoding='UTF-16') as f:
    f.write(cap.stdout)
```

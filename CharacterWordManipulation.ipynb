```python
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
english_chars = [chr(i) for i in range(ord('a') , ord('z')+1)]
#english_chars += [chr(i) for i in range(ord('0') , ord('9')+1)]
english_chars
```

```python
def count_chars(s , cs):
    res=0

    for c in s:
        res += 1 if c in cs else 0
    return res
```
